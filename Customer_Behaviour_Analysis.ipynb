

Review Rating has 37 null values .need to replace with either mean or median,
choose mean - will get affected by outliers
median -  robust to outliers

can work - but not best practice
median Review Rating within each category and will replace that with the missing value


```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

converting the columns to snake casing for easier handling

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

### Feature Engineering

```python
df.columns
```

```python
# Create a columns age_group
labels = ['Young Adult','Adult','Midlle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q = 4,labels = labels)
```

```python
df[['age','age_group']].head(10)
```

converting Frequency of Purchases to numerical form



```python
df['frequency_of_purchases'].unique()
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

both cols look the same as if a promo code used then a discount is applied
but sometimes businesses might apply discounts with out using promo code
like seaonal discounts , exslusive dicounts

now need to check do we need both the cols or one can be dropped?

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

both cols carry the same value - lets remove promo_code_used

```python
df = df.drop('promo_code_used', axis=1)
```
